# 反向传播

```python
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams['font.sans-serif'] = ['SimHei', 'Arial Unicode MS', 'DejaVu Sans']
plt.rcParams['axes.unicode_minus'] = False

np.random.seed(42)

n_samples = 100
areas_m2 = np.random.uniform(90, 150, n_samples)
areas = areas_m2 / 100
prices = 250 * areas + 30 + np.random.randn(n_samples) * 15

print(f"房屋数量: {len(areas)}")
print(f"面积范围: {areas.min():.2f} - {areas.max():.2f} 平方米")
print(f"价格范围: {prices.min():.2f} - {prices.max():.2f} 万元")
print(f"真实规律: price = 2.5 * area + 30")
```

```python
def gradient_descent(X, y, lr=0.1, epochs=200):
    w, b = 0.0, 0.0
    n = len(X)

    history = {'w': [w], 'b': [b], 'loss': [np.mean((w*X + b - y)**2)]}

    for epoch in range(epochs):
        y_pred = w * X + b
        dw = (2/n) * np.sum((y_pred - y) * X)
        db = (2/n) * np.sum(y_pred - y)

        w -= lr * dw
        b -= lr * db

        loss = np.mean((y_pred - y)**2)
        history['w'].append(w)
        history['b'].append(b)
        history['loss'].append(loss)

        if epoch % 20 == 0:
            print(f"Epoch {epoch}: w={w:.4f}, b={b:.4f}, loss={loss:.4f}")

    return history


history = gradient_descent(areas, prices, lr=0.01, epochs=100000)
print(f"最终参数: w={history['w'][-1]:.4f}, b={history['b'][-1]:.4f}")
```
